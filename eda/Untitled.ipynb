```python
import pandas as pd
import plotly.graph_objects as go

# Example data in Pandas DataFrame format
data = pd.DataFrame({
    'dates': ['2023-05-01', '2023-05-08', '2023-05-15', '2023-05-22', '2023-05-29'],
    'values': [100, 120, 80, 150, 130],
    'change_rates': [0, 20, -33.33, 87.5, -13.33]
})

# Create the bar chart trace
bar_trace = go.Bar(
    x=data['dates'],
    y=data['values'],
    name='Values'
)

# Create the line chart trace
line_trace = go.Scatter(
    x=data['dates'],
    y=data['change_rates'],
    name='Change Rates',
    yaxis='y2'
)

# Create the layout with two y-axes
layout = go.Layout(
    title='Weekly Values and Change Rates',
    xaxis=dict(title='Date'),
    yaxis=dict(title='Values'),
    yaxis2=dict(
        title='Change Rates (%)',
        overlaying='y',
        side='right',
        showgrid=False
    )
)

# Create the figure with both traces
fig = go.Figure(data=[bar_trace, line_trace], layout=layout)

# Show the chart
fig.show()
```
